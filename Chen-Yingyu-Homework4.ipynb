```python
import pandas as pd
import traitlets
import matplotlib.pyplot as plt
import ipywidgets
import numpy as np
```

```python
df = pd.read_csv('ufo.csv',
                 names = ["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "comment", "report_date", "latitude", "longitude"],
                  parse_dates = ["date", "report_date"])
df1 = df.fillna({'country':'unknown', 'state':'unknown', 'shape':'unknown'})
df1
```

```python
@ipywidgets.interact(x = ["date", "city", "state", "country",
                          "shape", "duration_seconds",
                          "report_date", "latitude", "longitude"],
                    y = ["date", "duration_seconds",
                          "report_date", "latitude", "longitude"],
                    style = plt.style.available)
def graph_builder(x, y,style):
    print('let us select at least a numerical column!')
    with plt.style.context(style):
        fig, ax = plt.subplots(1,1, figsize=(12,7))
        plt.scatter(df1[x], df1[y], 
            c=np.log10(df1['duration_seconds']))
    plt.show()
```

```python
@ipywidgets.interact(x = ["date", "city", "state", "country",
                          "shape"],
                    y = ["duration_seconds",
                          "report_date", "latitude", "longitude"],
                    num_of_bins= [5, 10, 15, 20, 50, 100, 150, 200, 500])
def binned_graph(x, y,num_of_bins):
    print('let us select at least a numerical column!')
    fig, ax = plt.subplots(figsize=(20,5))
    agg_data = df1.groupby(x)[y].sum()
    print(agg_data)
    agg_data.hist(bins = num_of_bins)
    #plt.bar(x,y,bins)
    #plt.hist(agg_data)
    plt.show()
```
